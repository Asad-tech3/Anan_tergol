```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import os
import json
from google.colab import drive

# 1. חיבור לדרייב
drive.mount('/content/drive')

# 2. הגדרת נתיב התיקייה והקובץ
FOLDER_NAME = '/content/drive/MyDrive/Anan'
FILE_PATH = f'{FOLDER_NAME}/students.json'

# נוודא שהתיקייה קיימת
os.makedirs(FOLDER_NAME, exist_ok=True)

# 3. אם הקובץ לא קיים - ניצור קובץ JSON התחלתי לדוגמה
if not os.path.exists(FILE_PATH):
    sample_students = [
        {
            "first_name": "Ashraf",
            "last_name": "Asadi",
            "email": "ashraf.asadi@e.braude.ac.il",
            "courses": ["Automata and Computability"],
            "interesting_link": "https://ashrfasadi.com"
        },
        {
            "first_name": "Asad",
            "last_name": "Rezek",
            "email": "assad.rezek@e.braude.ac.il",
            "courses": ["Web Computability"],
            "interesting_link": "https://example.com"
        },
        {
            "first_name": "Aya",
            "last_name": "Mohammed",
            "email": "Aya.mohamed@e.braude.ac.il",
            "courses": ["Web Computability"],
            "interesting_link": "https://example.com"
        },
        {
            "first_name": "Nora",
            "last_name": "Abood",
            "email": "Nora.Abood@e.braude.ac.il",
            "courses": ["Web Computability"],
            "interesting_link": "https://example.com"
        },
        {
            "first_name": "Marian",
            "last_name": "Sayegh",
            "email": "Marian.Sayegh@e.braude.ac.il",
            "courses": ["Web Computability"],
            "interesting_link": "https://example.com"
        }
    ]

    with open(FILE_PATH, 'w', encoding='utf-8') as f:
        json.dump(sample_students, f, ensure_ascii=False, indent=4)

# פונקציה לטעינת הסטודנטים מקובץ JSON
def load_students():
    if not os.path.exists(FILE_PATH):
        return []

    with open(FILE_PATH, 'r', encoding='utf-8') as f:
        return json.load(f)

students = load_students()

# נבנה מילון לפי שם מלא לצורך תצוגה נוחה
students_dict = {
    f"{student['first_name']} {student['last_name']}": student
    for student in students
}

# --- עיצוב ויצירת רכיבי הטופס ---
style = {'description_width': '140px'}
layout = widgets.Layout(width='500px', margin='6px 0')

title = widgets.HTML("""
<h2 style='color:#1a73e8; text-align:center; direction:rtl;'>
🎓 טופס ניהול סטודנטים - JSON
</h2>
""")

options_list = ['...בחר סטודנט'] + list(students_dict.keys())
dropdown = widgets.Dropdown(
    options=options_list,
    description='👤 בחר סטודנט:',
    style=style,
    layout=layout
)

# שדות להצגת פרטי הסטודנט
txt_first = widgets.Text(description='שם פרטי:', style=style, layout=layout, disabled=True)
txt_last = widgets.Text(description='שם משפחה:', style=style, layout=layout, disabled=True)
txt_email = widgets.Text(description='📧 מייל:', style=style, layout=layout, disabled=True)
txt_courses = widgets.Text(description='📚 קורסים:', style=style, layout=layout, disabled=True)
txt_link = widgets.Text(description='🔗 קישור מעניין:', style=style, layout=layout, disabled=True)

output_msg = widgets.Output()

# --- פונקציות פעולה ---
def on_student_select(change):
    if change['new'] != '...בחר סטודנט' and change['new'] in students_dict:
        data = students_dict[change['new']]
        txt_first.value = data.get('first_name', '')
        txt_last.value = data.get('last_name', '')
        txt_email.value = data.get('email', '')
        txt_courses.value = ", ".join(data.get('courses', []))
        txt_link.value = data.get('interesting_link', '')
    else:
        txt_first.value = ''
        txt_last.value = ''
        txt_email.value = ''
        txt_courses.value = ''
        txt_link.value = ''

dropdown.observe(on_student_select, names='value')

# --- עיצוב מסגרת לטופס ---
form_box = widgets.VBox([
    title,
    dropdown,
    txt_first,
    txt_last,
    txt_email,
    txt_courses,
    txt_link,
    output_msg
], layout=widgets.Layout(
    border='2px solid #1a73e8',
    padding='20px',
    width='550px',
    align_items='center'
))

# הצגת הטופס
display(form_box)
```
